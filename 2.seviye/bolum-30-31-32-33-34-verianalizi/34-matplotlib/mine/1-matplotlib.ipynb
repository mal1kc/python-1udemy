```python
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
```

```python
x = np.arange(1,6)
x
```

```python
y = np.arange(2,11,2)
y
```

```python
plt.plot(x,y,"red")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"blue")
plt.subplot(2,2,2)
plt.plot(y,x,"blue")
plt.subplot(2,2,3)
plt.plot(y,x,"blue")
plt.subplot(2,2,4)
plt.plot(x,x**2,"blue")
```

```python
x = np.arange(1,6)
y = np.arange(2,11,2)

fgr = plt.figure()

axes1 = fgr.add_axes([0.1,0.1,0.8,0.8])
axes2 = fgr.add_axes([0.2,0.5,0.3,0.3])

axes2.plot(x,y)
axes2.set_xlabel("inner X")
axes2.set_ylabel("inner y")
axes2.set_title("inner graph")


axes1.plot(y,x)
axes1.set_xlabel("outer X")
axes1.set_ylabel("outer y")
axes1.set_title("outer graph")
plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)

for ax in axes:
    ax.plot(x,y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)

axes[0].plot(x,y)
axes[0].set_title("first axes")
axes[1].set_title("second axes")
axes[1].plot(x,x**4)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(5,3))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple")
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,6))

axes[0].plot(x,y,color="red")
axes[0].set_title("axes 0")
axes[1].plot(x,x**4,color="red")
axes[1].set_title("axes 1")
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(6,3))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**0.5,color="red",label="x**0.5")
axes.plot(x,x**2,color="purple",label="x**2")
axes.plot(x,x**3,color="#4a386d",label="x**3")
axes.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,color="red",lw=3,ls="-.",marker="o",markersize=10,markerfacecolor="purple",markeredgecolor="black",mew=2.1)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,color="red",lw=3,ls="-.",marker="o",markersize=20,markerfacecolor="purple",markeredgecolor="black",mew=3)
axes.set_xlim(0,50)
axes.set_ylim(0,50)
```
