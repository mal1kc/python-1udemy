```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
x = np.arange(51)
y = np.arange(0,101,2)
z = x ** 3
```

# part 1

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="red")
axes.set_title("x vs y graph")
axes.set_xlabel("x values")
axes.set_ylabel("y values")
axes.set_xlim(0,50)
axes.set_ylim(0,100)
plt.show()
```

# part2 

```python
fig = plt.figure(figsize=(8,6))
axes1 = fig.add_axes([0,0.1,0.8,0.8])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])

axes1.plot(x,z,color="purple")
axes1.set_title("outer graph - x vs z graph")
axes1.set_xlim(0,50)
axes1.set_ylim(0,120000)

axes2.plot(x,y,color="red")
axes2.set_title("inner graph - x vs y graph")
axes2.set_xlim(0,50)
axes2.set_ylim(0,100)

plt.show()
```

# part 3

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(10,6))

axes[0].plot(x,y,color = "red",marker ="o",markersize=10
        ,markerfacecolor ="black")
axes[0].set_title("x vs y graph")
axes[0].set_xlabel("x values")
axes[0].set_ylabel("y values")
axes[0].set_xlim(0,50)
axes[0].set_ylim(0,100)

axes[1].plot(x,z,color = "purple",ls="-.")
axes[1].set_title("x vs z graph")
axes[0].set_xlabel("x values")
axes[0].set_ylabel("z values")
axes[1].set_xlim(0,50)
axes[1].set_ylim(0,120000)


plt.tight_layout()
plt.show()
```
